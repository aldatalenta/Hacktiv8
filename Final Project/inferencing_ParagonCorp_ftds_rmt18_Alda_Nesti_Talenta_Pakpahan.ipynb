# 2 - Import Library

```python
# Import Library
import pickle
import pandas as pd

# Import Library ARIMA
import pickle

# Library Visualisasi
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
```

# 3 - Import File Model

```python
# Load Model
with open('model_arima_res.pkl', 'rb') as file_1:
  model_arima_res = pickle.load(file_1)
```

# 4 - Forecasting of quantity in the next 12 weeks

```python
# Forecasting
forecast_next12weeks = model_arima_res.forecast(12)
forecast_next12weeks
```

```python
# Visualization of forecasting in the next 12 weeks
plt.figure(figsize=[20,10])
sns.lineplot(data=forecast_next12weeks)      # lineplot with seaborn
plt.title(f'Quantitiy All Products')     # set title
plt.xlabel('')
plt.show()
```
