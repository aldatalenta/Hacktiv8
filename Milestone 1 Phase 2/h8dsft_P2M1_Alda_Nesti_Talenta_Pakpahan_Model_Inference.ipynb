# x. Model Inference

>Model yang sudah dilatih akan dicoba pada data yang bukan termasuk ke dalam train-set, val-set, ataupun test-set. Data ini harus dalam format yang asli, bukan data yang sudah di-scaled.

Pada bagian ini akan dilakukan proses model inference dengan menggunakan data `inference set` dan melakukan preprocessing yang sama seperti yang dilakukan sebelumnya serta melakukan prediksi terhadap data `inference-set`.

## Import Libraries

```python
## REQUIRED LIBRARIES

# For load the files of model
import pickle

# For load model
from tensorflow.keras.models import load_model

# For data wrangling
import numpy as np  # linear algebra
import pandas as pd  # data processing
```

## Load model

```python
# Load All Files of Functional API (Improvement)

with open('final_pipeline.pkl', 'rb') as file_1:
  final_pipeline = pickle.load(file_1)

model_ann = load_model('churn_risk_model.h5')
```

## Load data

```python
# load data
df_inf = pd.read_csv('/content/data inference.csv')
df_inf
```

```python
# Transform Inference-Set

data_inf_transform = final_pipeline.transform(df_inf)
data_inf_transform
```

```python
# Predict using Artificial Neural Network

y_pred_inf = model_ann.predict(data_inf_transform)
y_pred_inf = np.where (y_pred_inf>=0.5, 1, 0)
y_pred_inf = pd.DataFrame(y_pred_inf, columns= ['churn_prediction'])
y_pred_inf
```

```python
# concate `churn_risk_score` in data `df_inf` and the result of `y_pred_inf`
y = df_inf['churn_risk_score']
df_inf_final = pd.concat([y, y_pred_inf], axis=1)
df_inf_final
```

Berdasarkan dataFrame di atas, kita dapat mengetahui bagaimana keakuratan model memprediksi apakah seorang customer akan berhenti memakai produk yang ditawarkan atau tidak. Kita dapat melihat bahwa tidak ada perbedaan antara hasil prediksi dengan data aktual. Artinya, model ini sudah bagus untuk memprediksi seorang custumer akan churn atau tidak dan bisa digunakan oleh perusahaan untuk meminimalisir resiko customer churn.<br>
Hal ini dapat dibuktikan juga pada proses evaluasi model sebelumnya, kita mendapatkan bahwa model ini memiliki performansi terbaik mencapai `akurasi 92%` setelah dilakukan improvement.
